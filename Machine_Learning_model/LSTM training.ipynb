```python
def create_sequences_with_ids(df, window, feat_cols, target_cols):
    X_seq, y_seq, obj_ids, indices = [], [], [], []
    for obj_id, group in df.groupby('object_id'):
        group = group.sort_values('pointEndTime')
        feats = group[feat_cols].values
        targets = group[target_cols].values
        codes = group['object_code'].values

        for i in range(len(group) - window):
            X_seq.append(feats[i:i+window])
            y_seq.append(targets[i+window])
            obj_ids.append(codes[i+window])
            indices.append(group.index[i+window])  # keep the row index
    return np.array(X_seq), np.array(y_seq), np.array(obj_ids), np.array(indices)
```

```python
new_space_data['dt'] = new_space_data.groupby('object_id')['pointEndTime'].diff().dt.total_seconds()
new_space_data['dt'] = new_space_data['dt'].fillna(0)

# Encode object_id for embedding
le = LabelEncoder()
new_space_data['object_code'] = le.fit_transform(new_space_data['object_id'])

# Features and targets
feat_cols = ['x','y','z','vx','vy','vz','dt']
target_cols = ['x','y','z']

# -----------------------------
# 2. Create sequences grouped by object_id
# -----------------------------
def create_sequences_with_ids(df, window, feat_cols, target_cols):
    X_seq, y_seq, obj_ids, indices = [], [], [], []
    for obj_id, group in df.groupby('object_id'):
        group = group.sort_values('pointEndTime')
        feats = group[feat_cols].values
        targets = group[target_cols].values
        codes = group['object_code'].values

        for i in range(len(group) - window):
            X_seq.append(feats[i:i+window])
            y_seq.append(targets[i+window])
            obj_ids.append(codes[i+window])
            indices.append(group.index[i+window])  # keep the row index
    return np.array(X_seq), np.array(y_seq), np.array(obj_ids), np.array(indices)

window_size = 20
X, y, obj_ids, row_indices = create_sequences_with_ids(new_space_data, window_size, feat_cols, target_cols)
print("X shape:", X.shape, "y shape:", y.shape, "obj_ids shape:", obj_ids.shape)

# -----------------------------
# 3. Scale features
# -----------------------------
nsamples, ntimesteps, nfeatures = X.shape

# -----------------------------
# 4. Train/validation split
# -----------------------------
X_train, X_val, y_train, y_val, ids_train, ids_val, idx_train, idx_val= train_test_split(
    X, y, obj_ids, row_indices, test_size=0.2, shuffle=True, random_state=42
)

# -----------------------------
# 5. Build LSTM model with object_id embedding
# -----------------------------
num_objects = len(le.classes_)
embed_dim = 8  # embedding dimension for object_id

# Sequence input
seq_input = Input(shape=(window_size, nfeatures))

# Object_id input
id_input = Input(shape=(1,))
id_embed = Embedding(input_dim=num_objects, output_dim=embed_dim)(id_input)
id_embed = Flatten()(id_embed)

# LSTM branch
x = LSTM(128, return_sequences=True)(seq_input)
x = Dropout(0.2)(x)
x = LSTM(64)(x)

# Concatenate LSTM output with object_id embedding
x = Concatenate()([x, id_embed])
x = Dense(64, activation='relu')(x)
output = Dense(3)(x)  # next x,y,z

model = Model(inputs=[seq_input, id_input], outputs=output)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')

# -----------------------------
# 6. Train model with EarlyStopping
# -----------------------------
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history = model.fit(
    [X_train, ids_train], y_train,
    validation_data=([X_val, ids_val], y_val),
    epochs=100,
    batch_size=64,
    callbacks=[early_stop],
    verbose=1
)
```
